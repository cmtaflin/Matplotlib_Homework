```python

# Import Libraries and set up canvas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
##Import Data
##City Data
city_data_csv=("city_data.csv")
city_data=pd.read_csv(city_data_csv)

ride_data_csv=("ride_data.csv")
ride_data=pd.read_csv(ride_data_csv)

data_merged=pd.merge(city_data,ride_data,on="city")        
```

                 SCATTER PLOT QUESTIONS

```python
#Divide Data into Types
urban_data=data_merged[data_merged["type"]=="Urban"]
suburb_data=data_merged[data_merged["type"]=="Suburban"]
rural_data=data_merged[data_merged["type"]=="Rural"]

### Urban Metrics
u_avg_fare=urban_data.groupby(["city"]).mean()['fare']
u_rides_per_city=urban_data.groupby(["city"]).count()['ride_id']
u_total_drivers=urban_data.groupby(["city"]).count()['ride_id']

### Suburb Metrics
s_avg_fare=suburb_data.groupby(["city"]).mean()['fare']
s_rides_per_city=suburb_data.groupby(["city"]).count()['ride_id']
s_total_drivers=suburb_data.groupby(["city"]).count()['ride_id']

### Rural Metrics
r_avg_fare=rural_data.groupby(["city"]).mean()['fare']
r_rides_per_city=rural_data.groupby(["city"]).count()['ride_id']
r_total_drivers=rural_data.groupby(["city"]).count()['ride_id']
```

```python
### Build my scatter plan #####
plt.scatter(u_rides_per_city,u_avg_fare,s=15*u_rides_per_city,marker= "o", linewidth=1.0, facecolor="coral", \
            edgecolors="black",Label="Urban")
plt.scatter(s_rides_per_city,s_avg_fare,s=12*s_rides_per_city,marker= "o", linewidth=1.0, facecolor="lightblue", \
            edgecolors="black",Label="Suburb")
plt.scatter(r_rides_per_city,r_avg_fare,s=10
            *r_rides_per_city,marker= "o", linewidth=1.0, facecolor="yellow", \
            edgecolors="black",Label="Rural")

#Title and Axis
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare (Dollares)")

#Grid lines and Limits
plt.xlim ((0,41))
plt.grid(True)

City_Types= ["Urban","Suburb","Rural"]

#Create a Legend
legend=plt.legend(fontsize="small",title="City_Types")

legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]

#Grid lines and Limits
plt.xlim ((0,41))
plt.grid(True)

# Incorporate a text label regarding circle size

t = (" Note: Circle size correlates with driver count by city")
plt.text(58, 35, t, ha='left', horizontalalignment='center', verticalalignment='center')

# Save Figure
#plt.savefig("../Images/Pyberrideshare.png")
# Show plot
plt.show()
```

                                Total Fares By Type

```python
#collect city type data (desired dataset to be analyzed)
u_percent_city_type=data_merged.groupby(["type"]).sum()["fare"]

#Build Pie Chart items

city_types=["Rural","Suburban","Urban"]
colors = ["Yellow","Lightblue","Coral"]

plt.pie(u_percent_city_type, labels=city_types,explode=(0,0,0.1),colors=colors, \
        autopct="%1.1f%%", shadow=False, startangle=140)
plt.title("% Total Fares by City Type")
plt.show()
#Save Figure
```
